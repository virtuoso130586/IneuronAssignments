```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
df = pd.read_csv(url)
df.head()
```

## Create a pie chart presenting the male/female proportion

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
sex_labels = 'Male', 'Female'
df["sex"].value_counts()
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(df["sex"].value_counts(), labels=sex_labels,shadow=True, startangle=90, autopct='%1.1f%%')
plt.show()
```

## 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.rcParams["figure.figsize"] = (20,10)
male_plot = df[df.sex=='male'].plot.scatter('age','fare', color='green', label="Male")
df[df.sex=='female'].plot.scatter('age','fare', color='red', label="Female", ax=male_plot)
plt.figure()
```
